```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Sales-Superstore.csv')
df.info()
```

```python
df['Category'].value_counts().plot(kind = 'bar', label = 'Category', color = 'green')
plt.legend()
```

```python
df['Segment'].value_counts().plot(kind = 'bar', label = 'Segment', color = 'blue')
plt.legend()
```

```python
#df['Sub-Category'].value_counts().plot(kind = 'bar', color = 'purple')
#plt.legend()

dff = df.loc[df['Segment'] == 'Corporate']
dff.info()
```

```python
dff['Region'].value_counts().plot(kind = 'bar', label = 'Category', color = 'purple')
plt.legend()
```

```python
dfff = dff.loc[dff['Region'] == 'West']
dfff.info()
```

```python
dfff['State'].value_counts().plot(kind = 'bar', color = 'pink')
plt.legend()
```

```python
final = dfff.loc[dfff['State'] == 'California']
final.info()
```

```python
cols = ['Segment', 'Region', 'State', 'Country']

final = final.drop(cols, axis = 1)
final.info()
```

```python
final0 = final.loc[final['Category'] == 'Office Supplies']
final0.describe().T
```

```python
final0['X'] = (final0['Sales'])/(final0['Quantity'])
final0.describe()
```
